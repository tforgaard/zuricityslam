```python
import networkx as nx

from pathlib import Path

from cityslam.utils.parsers import find_models
from cityslam.utils.graph import find_graphs, get_graphs, draw_super, draw_graphs, transform_models, get_tf_filter_view
```

```python
# Setup the paths
base_dir = Path('/cluster/project/infk/courses/252-0579-00L/group07')

models = base_dir / 'outputs' / 'models-features'
graphs = base_dir / 'outputs' / 'merge'
outputs = base_dir / 'outputs' / 'graph'

outputs.mkdir(exist_ok=True, parents=True)
```

```python
super_graph = find_graphs(models, graphs)
maps = get_graphs(super_graph)
print(f"Number of nodes: {len(super_graph.nodes)}")
```

```python
draw_super(get_tf_filter_view(super_graph), models)

draw_graphs(maps[:7])
```

```python
import matplotlib.pyplot as plt

from natsort import natsorted

A = nx.to_numpy_array(get_tf_filter_view(super_graph),nodelist=natsorted(super_graph.nodes))

plt.imshow(A,interpolation=None)
```
